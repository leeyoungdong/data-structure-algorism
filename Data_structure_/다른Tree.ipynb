# **Red-Black Tree**

레드-블랙 트리는 이진 탐색트리(Binary Search Tree)의 한종류.

삽입, 삭제, 검색 연산을 O(log n)으로 수행하도록 보장하는 균형잡힌 트리를 말한다.
즉, 트리의 높이가 log n 이 되도록 한다.

image.png

레드-블랙트리의 조건

* 모든 노드는 빨간색 or 검은색
* 루트 노드는 검은색
* Null 혹은 Nil로 표기된 리프 노드는 검정색
* 빨간색 노드의 자식 노드는 검정색이다. 즉, 빨간색 노드가 연속적으로 나올 수 없다.
* 리프노드에서 루트노드까지 가는 경로에서 만나는 검은색 노드의 개수는 같다.


**레드-블랙트리가 균형 잡힌 트리인 이유**

레드-블랙 트리의 5번째 조건때문

검은색 노드의 개수가 B, 빨간색 노드가 최소가 되는 경우와 최대가 되는 경우를 고려해본다면, 

빨간색 노드가 최소가 되려면
* 빨간색 노드 = 0개
* 총 노드의 개수 = B개

빨간색 노드가 최대가 되려면
* 검-빨-검-빨..로 반복되어야 한다.
* 총 노드의 개수 = 2B개

그러므로 최소 경로와 최대 경로 차이가 2배보다 크지 않으므로 레드-블랙 트리는 균형잡힌 트리라고 말할 수 있다.

레드-블랙 트리의 연산 - 트리와 동일 / 검색,삽입,제거

# **B-Tree**

B-트리는 이진 트리(BInary Tree)를 확장해 모든 리프 노드들이 같은 높이를 갖도록 하는 트리.

노드 내에 여러개의 key가 있을 수 있으며, 최대 key의 개수에 따라 2개이면 2차 B-트리, N개면 N차 B-트리리

image.png

**B-트리의 조건**

* 노드의 key개수가 N이면, 자식의 노드개수는 N+1개
* 노드 내의 key는 오름차순으로 정렬
* 루트 노드는 2개 이상의 자식을 가져가야함
* 루트 노드를 제외한 나머지 노드들은 적어도 최대 M/2개의 key를 가져가야함(M은 B-트리의 차수)
* 리프 노드는 모두 같은 레벨에 있어야함.

**연산**

B-트리의 연산 - 트리와 동일 / 검색,삽입,제거

**B-트리 vs B+ 트리**

B+트리는 B-트리와 비슷하지만, 리프노드가 연결리스트의 형태를 띄어 선형 검색이 가능한 트리

모든 노드에 key와 data가 있는 B트리와 달리 B+트리는 리프노드에만 data가 존재. 또한 삽입과 제거 연산 모두 리프 노드에서만 이뤄짐
