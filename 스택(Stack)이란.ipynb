# **stack이란 ?**

image.png

LIFO (Last In First Out) 또는 FILO(First In, Last Out) 구조의 자료형으로 한 쪽으로만 데이터를 넣고 뺄 수 있다. push 명령으로 데이터를 넣고, pop 명령으로 가장 마지막에 들어간 데이터를 빼낸다.

[데이터를 제한적으로 접근할 수 있는 구조이다. 한 쪽 끝에서만 자료를 넣거나 뺄 수 있다.]

stack 은 브라우저의 뒤로가기 기능, ctrl + z (되돌리기), 지역 변수와 매개변수를 저장하는 stack 메모리 등에 사용된다. 이외에도 DFS 알고리즘 등 다양한 곳에 사용되는 자료형이다.

[컴퓨터 내부의 프로세스 구조의 함수 동작 방식에서 사용된다.]

stack 에 데이터가 꽉 차서 더 넣을 공간이 없는데 데이터를 push 하는 경우 overflow, 반대로 데이터가 없는데 pop 하는 경우를 underflow 라고 한다.

**장점**

구조가 단순해서 구현이 쉽다.

데이터 저장 및 읽기의 속도가 빠르다.

**단점**

데이터 최대 갯수를 미리 정해야한다.

파이썬의 경우 재귀 함수는 1000번까지만 호출이 가능하다.

저장 공간의 낭비가 발생할 수 있다.

미리 최대 갯수만큼 저장 공간을 확보해야한다.

스택은 단순하고 빠른 성능을 위해 사용되므로 보통 배열 구조를 활용하여 구현하는 것이 일반적이다. 이 경우, 위에서 열거한 단점이 있을 수 있다.

# **리스트로 스택 구현**

```python
stack_list = list()

def push(data):
  stack_list.append(data)

def pop():
  data = stack_list[-1]
  del stack_list[-1]
  return data
```

```python
for i in range(15):
  push(i)
  print(stack_list)
```

```python
for i in range(5):
  pop()
  print(stack_list)
```
