# **Heap(binary heap)**

최대값 및 최소값을 찾아내는 연산을 빠르게 하기 위해 고안된 완전 이진 트리를 기본으로하는 자료구조로, 다음의 속성을 만족

* A가B의 부모노드이면, A의 키값과 B의 키값사이에는 대소관계가 성립

heap의 종류에는 min / max heap이 있다.

각 노드의 자식 노드의 최대 개수는 힙의 종류에 따라 상이, 대부분의 경우는 자식 노드의 개수가 최대 2개인 이진 힙(binary heap)을 사용

힙에서는 가장 높은(or 가장 낮은) 우선 순위를 가지는 노드가 항상 루트 노드에 오게 되는 특징이 있으며, 이를 응용하여 우선순위 큐와 같은 추상적 자료형을 구현할 수 있다.

**이진 힙(binary heap)**

이진 힙은 다음과 같은 두가지 특징을 갖는다. 트리를 T, 임의 내부 노드를 V라고하면 다음과 같다.

* 루트 노드를 제외한 각 내부 노드는 key(T.parent(V)) < key(V) or key(T.parent(V)) > key(V)이다 (즉, 키 값은 오름차순 OR 내림차순)
* 마지막 왼쪽 결함 노드들의 레벨을 제외한 다른 모든 레벨들은 완전 이진진 트리를 형성한다

힙 리스트(heap list)로 표현할 때 i번째 노드의 왼족 자식 노드의 위치는 2i가 되며, i번째 노드의 오른쪽 자식 노드의 위치는 2i+1이고, 또한 i번째 노드의부모 노드의 위치는 1/2가 된다.

이진 힙의 시간 복잡도는 O(log n)이다.

# **Min heap**

image.png

최소 힙(Min heap)은 부모 노드이 키 값이 자식 노드의 키 값보다 작거나 같은 완전 이진 트리아다.

key(부모노드) <= key(자식노드)

# **Max heap**

image.png

최대 힙(max heap)은 부모 노드의 키 값이 자식 노드의 키 값보다 크거나 같은 완전 이진 트리아다.

key(부모노드) >= key(자식노드)
