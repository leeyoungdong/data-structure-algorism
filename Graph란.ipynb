# **Graph란?**

그래프는 정점과 간선으로 이뤄진 자료구조. 정점간의 연결관계는 간선으로 나타냄

**그래프의 종류**

image.png

간선이 담고있는 정보와 연결 상태에 따라 그래프의 종류가 나뉨.

 정점을 연결하는 간선에 방향이 없다 - 무방향 그래프 

 정점을 연결하는 간선에 방향이 존재 - 방향 그래프(방향그래프는 간선의 방향으로만 이동할수 있다)

 두 정점을 이동할 때 비용이 발생하면 가중치 그래프로 나타낼 수 있다.

모든 정점이 간선으로 연결된 경우 - 완전 그래프

**그래프 구현 방식**

인접 행렬 방식 - 노드를 인덱스로 삼는 2차원 배열을 만들어 각 노드가 간선으로 연결되어 있으면 배열에 1을 넣어주고, 연결되지 않았다면 0을 넣어줌

두 노드의 연결관계를 조회할때  O(1)의 시간이 걸림, 그러나 모든 정점에 대해, 간선 정보를 입력해야 하므로 초기화에는 O(N^2)의 시간이 소요됨

노트의 수가 많고, 간선의 수가 적은 그래프의 경우, 공간을 낭비하게 됨

image.png

인접 리스트 방식 - 그래프의 노드들을 리스트로 표현한다. HEAD노드와 연결된 노드들을 링크에 달아주면 됨

정점에 연결된 노드들의 정보를 얻기 위해서 O(N)의 시간이 걸림. (N-간선의 수) 간선 정보만 유지하므로, 공간 낭비가 적으나 두 정점이 연결되어있는지 확인하기 위해서 O(M)시간이 걸리며, 구현이 비교적 어려움

**그래프 용어**

* 정점(Vertice) : 노드(node)와 동일어 정점에는 데이터가 저장됨
* 간선(edge) : 링크(arcs)라고도 하며 노드간의 관계를 나타냄
* 인접 정점(adjacent vertex): 간선에 의해 직접 연결된 정점
* 단순 경로(simple-path) : 경로 중 반복되는 정점이 없는것, 같은 간선을지나가지 않는 경로
* 차수(degree): 무방향 그래프에서 하나의 정점에 인접한 정점의 수
* 진출 차수(out-degree) : 방향그래프에서 사용되는 용어로 한 노드에서 외부로 향하는 간선의 수
* 진입 차수(in-degree) : 방향그래프에서 사용되는 용어로 외부 노두에서 들어오는 간선의 수수
